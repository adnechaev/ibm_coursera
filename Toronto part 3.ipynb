## Part 1 - Formation of DataSet

Step 1 - Let's import libraries which are necessary for our work:

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import json
import requests
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
from bs4 import BeautifulSoup
print('Libraries imported.')
```

Step 2 - Let's download Wikipedia page with Candaian Postal Codes:

```python
wiki = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(wiki,'xml')
table = soup.find('table')
table
```

Step 3 - Let's work with the table:

Step 3.1 - Let's create table based on td tag.

```python
table_1 = table('td')
data_1 = []
for row in table_1:
    td=[]
    for t in row('p'):
        td.append(t.text.strip())
        data_1.append(td)
df1 = pd.DataFrame(data_1, columns=['Column 1'])
df1.head(20)
```

Step 3.2 - We can see that the data orginized inproper due to bad html code in wikipedia. So we need another tag. For postal code it could be b tag but for another data there are no proper tag. We will use span tag.

```python
PostalCode=[]
for row in table_1:
    for t in row('b'):
        PostalCode.append(t.text.strip())
print(PostalCode)
```

```python
Borough=[]
for row in table_1:
    for t in row('span'):
        Borough.append(t.text.strip())
print(Borough)
```

Step 3.3 - As we can see we have a good list for Postal Code and very bad and inproper for Boroughs and Neighborhoods. We need to split the items of this list on the basis of Borough or Neighborhood. It's possible on the basis of ( symbol. But before we mast be sure that we could recieve 'Not assigned' become 'Not assigned('.

```python
Borough2=[]
for i in Borough:
    if 'Not assigned' in i:
        i1='Not assigned('
        Borough2.append(i1)        
    else:
        i2=i
        Borough2.append(i2)
print(Borough2)
```

Step 3.4 - Morover, there are some items which consists of improper information. Let's find thier indexes and then correct it.

```python
print(Borough2.index('East TorontoBusiness reply mail Processing Centre969 Eastern(Enclave of M4L)'))
print(Borough2.index('Downtown TorontoStn A PO Boxes25 The Esplanade(Enclave of M5E)'))
print(Borough2.index('MississaugaCanada Post Gateway Processing Centre(Enclave of L4W)'))
print(Borough2.index('Queen\'s Park / Ontario Provincial Government'))
print(Borough2.index('EtobicokeNorthwest(Clairville / Humberwood / Woodbine Downs / West Humber / Kipling Heights / Rexdale / Elms / Tandridge / Old Rexdale)'))
print(Borough2.index('North York(Downsview)East  (CFB Toronto)'))
print(Borough2.index('North York(Don Mills)South(Flemingdon Park)'))
print(Borough2.index('East YorkEast Toronto(The Danforth  East)'))
```

Step 3.5 - Information became more clear and it become possible to split it on the basis of ( symbol.

```python
Borough3=[]
for i in Borough2:
    Borough3.append(i.rsplit('('))
print(Borough3)
```

Step 3.6 - The information become much better. We can create DataFrame to see how it looks like. I used column 'Neighbor' instead of recomended 'Neghborhoods' because we will work with it in the future.

Step 3.7 - Let's extract 'Neighbor' column from this dataframe to list.

Step 3.8 - Let's preprocess this list. Let's delete last ) symbol by splitting, creating new dataframe and extrcacting 'Neighbor 2' column.

```python
Neighbor2=[]
for i in Neighbor:
    Neighbor2.append(i.rsplit(')'))
print(Neighbor2)
```

Step 3.9 - Let's delete / symbol and use , symbol for our Neghborhoods. And let's create DataFrame of Neighborhoods.

```python
Neighbor4=[]
for i in Neighbor3:
    Neighbor4.append(i.rsplit(' / '))
print(Neighbor4)
```

```python
Neighbor6=[]
s=', '
for i in Neighbor4:
    for j in i:
        Neighbor5=[] 
        s1=s.join(i)
        Neighbor5.append(s1)
    Neighbor6.append(Neighbor5)
print(Neighbor6)
```

## Part 3 - Vizualizing and Clustering

Step 1 - Let's import libraries.

Step 2 = Let's create a map of Toronto

```python
address='Toronto, Canada'
geolocator=Nominatim(user_agent='to_explorer')
location=geolocator.geocode(address)
latitude=location.latitude
longitude=location.longitude
map_toronto=folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

Step 3 - Let's create all Post Codes on the map of Toronto

```python
for lat, lng, borough, post in zip(df9['Latitude'], df9['Longitude'], df9['Borough'], df9['PostalCode']):
    label = '{}, {}'.format(borough, post)
    label = folium.Popup(label)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue').add_to(map_toronto)  
    
map_toronto
```

Step 5 - Let's see which PostCodes are for each borough.

```python
toronto_onehot = pd.get_dummies(df9[['Borough']], prefix="", prefix_sep="")

toronto_onehot['PostalCode'] = df9['PostalCode'] 

fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot
```

```python
kclusters = 11

toronto_grouped_clustering = toronto_onehot.drop('PostalCode', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

labelsK=pd.DataFrame(kmeans.labels_, columns=['Cluster'])
kmeans.labels_
```

```python
toronto_clust=df9.join(labelsK)
toronto_clust
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(toronto_clust['Latitude'], toronto_clust['Longitude'], toronto_clust['Borough'], toronto_clust['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster))
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1]).add_to(map_clusters)
       
map_clusters
```

So we see on the map how Toronto is divided and which parts of Toronto has more or less Post Codes.
